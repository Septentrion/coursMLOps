```python
import pandas as pd
bitcoins = pd.read_csv('../data/Cours_BTC2FEUR_-_Feuille_1.csv')
bitcoins.head()
```

```python
# bitcoins['Date'] = pd.to_datetime(bitcoins['Date'])

bitcoins['Prix'] = bitcoins['Prix'].str.replace('1 BTC = ', '')
bitcoins['Prix'] = bitcoins['Prix'].str.replace('€', '').astype(float)
```

```python
bitcoins.dtypes
```

```python
bitcoins['Date'] = pd.to_datetime(bitcoins['Date'])
bitcoins.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

bitcoins.plot(x= 'Date',y ='Prix', rot = 90, figsize = (25,25))
plt.show()
```
